```python
from PIL import Image
import numpy as np

# Load your original image and the mask
image = Image.open('/content/drive/MyDrive/Inpainting/tt.jpg').convert('RGB')  # Replace with your filename
mask = Image.open('/content/drive/MyDrive/Inpainting/mask.png').convert('L')      # Replace with your mask file
```

```python
import torch
print(torch.cuda.is_available())  # Should output True
print(torch.cuda.current_device())
print(torch.cuda.get_device_name(torch.cuda.current_device()))
```

```python
import torch
from diffusers import StableDiffusionInpaintPipeline

pipe = StableDiffusionInpaintPipeline.from_pretrained(
    "runwayml/stable-diffusion-inpainting", torch_dtype=torch.float16
).to("cuda")

prompt = "A baby girl portrait, photo realistic, studio lighting, natural look"

result = pipe(prompt=prompt, image=image, mask_image=mask).images[0]
result.save("/content/drive/MyDrive/Inpainting/restored.jpg")
```
